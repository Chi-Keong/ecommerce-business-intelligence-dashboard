```python
# Checking nulls and duplicates
files = [customers, orderItems, orders, payments, products]
customers.attrs['name'] = "customers"
orderItems.attrs['name'] = "orderItems"
orders.attrs['name'] = "orders"
payments.attrs['name'] = "payments"
products.attrs['name'] = "products"
for file in files:
    na = file.isnull().sum()
    dup = file.duplicated().sum()
    print(f"Checking if Nulls are in dataset {file.attrs['name']}: \n", na, f"\n Checking if Duplicates are in dataset {file.attrs['name']}: \n", dup)
```

```python
# Removing the deplicates
products_noDuplicates = products.drop_duplicates()
```

```python
# Checking dataset products after cleaning
products_noDuplicates.info()
```

```python
# Converting date column to datatime for analysis
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
orders["order_approved_at"] = pd.to_datetime(orders["order_approved_at"])
orders["order_delivered_timestamp"] = pd.to_datetime(orders["order_delivered_timestamp"])
orders["order_estimated_delivery_date"] = pd.to_datetime(orders["order_estimated_delivery_date"])
```

```python
# Checking if the dates are valid according to business logics
#   -order purchase time should be later than approval time
orders[orders["order_purchase_timestamp"] > orders["order_approved_at"]]
```

```python
# Checking if the dates are valid according to business logics
#   -order purchase time should be later than actual delivery time
orders[orders["order_purchase_timestamp"] > orders["order_delivered_timestamp"]]
```

```python
# Checking if the dates are valid according to business logics
#   -order purchase time should be later than estimated delivery time
orders[orders["order_purchase_timestamp"] > orders["order_estimated_delivery_date"]]
```

```python
# Feature engineering
# Creating column of actual delivery days
orders["actual_delivery_days"] = (orders['order_delivered_timestamp'] - orders["order_purchase_timestamp"]).dt.days
```

```python
# Feature engineering
# Creating column of estimated delivery days
orders["expected_delivery_days"] = (orders['order_estimated_delivery_date'] - orders["order_purchase_timestamp"]).dt.days
```

```python
# Feature engineering
# labelling order which actual delivery time is later than estimated delivery time
orders["is_late"] = orders["order_delivered_timestamp"] > orders["order_estimated_delivery_date"]
```
